```python
# Subset db by session

def session_subset(df, session):
    subset_df = df.loc[df['session'] == session]
    return subset_df

# Count values per run in given dataframe

def count_values(df):
    val_counts = {}
    for run in df.columns[2:]:
        val_count = {'bids (tier1)': 0, 'bids (s3)': 0, 'no bids': 0, 'delete (tier1)': 0, 'delete (s3)': 0, 'NaN': 0}
        for idx, name in enumerate(df[run].value_counts().index.tolist()):
            count = df[run].value_counts()[idx]
            val_count[name] = count
        nans_count = df[run].isna().sum()
        val_count['NaN'] = nans_count
        val_counts[run] = val_count
    return pd.DataFrame.from_dict(val_counts)
```

```python
year1_df = session_subset(bids_db, 'ses-baselineYear1Arm1')
year1_value_counts = count_values(year1_df)
for re in ['T1.*', 'T2.*', '.*rest.*', '.*nback.*', '.*SST.*', '.*MID.*']:
    display(year1_value_counts.filter(regex=re))
```

```python
# Check for fully unprocessed subjects

def query_unprocessed_subjects(df):
    # Filter dataframe to get dataframe of subjects that are missing one or more modalities
    missing_data_df = df[df[df.columns[2:]].isin(['no bids']).any(axis=1)]
    # Filter again to remove subjects that have BIDS data somewhere
    fully_unprocessed_df = missing_data_df[~missing_data_df.isin(['bids (tier1)', 'delete (tier1)', 'bids (s3)', 'delete (s3)']).any(axis=1)]
    
    return fully_unprocessed_df

def query_processed_subjects(df):
    # Filter dataframe to get dataframe of subjects that do not have any unprocessed images
    processed_df = df[~df[df.columns[2:]].isin(['no bids']).any(axis=1)]
    # Filter again to remove subjects that have need data deleted
    fully_processed_df = processed_df[~processed_df.isin(['delete (tier1)', 'delete (s3)']).any(axis=1)]
    # Fulter again to remove subjects that do not have a T1
    fully_processed_df = query_hasT1(fully_processed_df)
    return fully_processed_df
```

```python
processed_df = year1_df[~year1_df[year1_df.columns[2:]].isin(['no bids']).any(axis=1)]
fully_processed_df = processed_df[~processed_df.isin(['delete (tier1)', 'delete (s3)']).any(axis=1)]
fully_processed_df = query_hasT1(fully_processed_df)
fully_processed_df
```
